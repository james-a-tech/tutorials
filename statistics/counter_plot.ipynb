Contour plot is used for visualization of 3D objects into 2D dimentions [Youtube explanation](https://youtu.be/WsZj5Rb6do8?si=If7ECNtMUfOxQMYv)

The below is used to exploring counter plot and its respective 3D visulization 


```python
import numpy as np
x, y = np.meshgrid(np.arange(0, 10, 0.1), np.arange(0, 10, 0.1))
# modify this function to get different types of outputs
h = 1 / (1 + np.exp(-(x - 5) ** 2 - (y - 5) ** 2))

# plot in counter plot matplot lib
import matplotlib.pyplot as plt
plt.contourf(x, y, h)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour Plot')
plt.show()

import plotly.graph_objects as go
# plot the same data in 3d plotly
fig = go.Figure(data=[go.Surface( x=x, y=y,z=h)])
fig.update_layout(title='3D Contour Plot', autosize=False,
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90))
fig.show()

```
